```python
data2 = pd.read_excel('sdgexcel.xlsx', sheet_name='Sheet2')
plt.figure(figsize=(30,12)) 
plt.title('District-wise (%)population with Moderately Elevated Blood Pressure\n\n', fontweight='bold',fontsize=25)
plt.xlabel('Districts', fontweight ='bold', fontsize = 25)
plt.ylabel('Moderately or severely elevated blood pressure\n(Systolic ≥160mm of Hg and/or Diastolic ≥100mm of Hg) (%)',fontweight ='bold', fontsize = 25)

plt.ylim(2,13)


df3 = pd.DataFrame(data2, columns=['Districts','93. Moderately or severely elevated blood pressure (Systolic ≥160mm of Hg and/or Diastolic ≥100mm of Hg) (%)'])
x = df3['Districts']
y = df3['93. Moderately or severely elevated blood pressure (Systolic ≥160mm of Hg and/or Diastolic ≥100mm of Hg) (%)']
plots = plt.plot(x, y, color='red', marker='o',label='Women- 15 and above years')


df4 = pd.DataFrame(data2, columns=['Districts', '96. Moderately or severely elevated blood pressure (Systolic ≥160mm of Hg and/or Diastolic ≥100mm of Hg) (%)'])
x = df4['Districts'] 
y = df4['96. Moderately or severely elevated blood pressure (Systolic ≥160mm of Hg and/or Diastolic ≥100mm of Hg) (%)']
plots = plt.plot(x, y, color='blue', marker='*', label='Men- 15 and above years')

plt.grid(True)
plt.xticks(rotation=90)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(framealpha=1, frameon=True, loc ="upper right",prop={"size":18});


plt.savefig('District-wise (%)population with Moderately Elevated Blood Pressure.jpg',dpi=300, bbox_inches='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt


fig = plt.subplots(figsize =(10, 10))

X = ['Mildly elevated\nblood pressure','Moderately or severely\nelevated blood pressure','Elevated\nblood pressure ']
Female = [14.3, 6.3, 29.1]
Male = [20.5, 9.1, 36.8]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.125, Female, 0.25, label = 'Women')
plt.bar(X_axis + 0.125, Male, 0.25, label = 'Men')

for i in range(len(Female)):
    plt.text(i-0.25, Female[i]+0.3, str(Female[i]))
    
for i in range(len(Male)):
    plt.text(i, Male[i]+0.3, str(Male[i]))

plt.xticks(X_axis, X)
plt.xlabel('\nBlood Pressures', fontsize = 20)
plt.ylabel('(%) population in Urban Areas\n', fontsize = 20)
plt.title("Comparision between Blood Pressure of Males and Females in Urban Areas\n\n", fontsize = 22)

plt.legend(borderpad=1)
plt.grid(axis = 'y', color = 'gray', linestyle = '--', linewidth = 0.25)

plt.savefig("Comparision between Blood Pressure of Males and Females in Urban Areas.jpg", dpi=300, bbox_inches='tight')
```
